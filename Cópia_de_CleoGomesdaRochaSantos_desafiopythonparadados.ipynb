3. Das pessoas que atuam com engenharia de software qual a
porcentagem de obesos?

```python
#filtrando o dataframe para engenheiros de software
engsoft = df[df['Profissão'] == 'Eng. de Software']

#frequência absoluta IMC
fr_absoluta_engsoftob = pd.crosstab(engsoft['Profissão'], engsoft['Categoria IMC'])

#aplicação da fórmula
pc_engsoftob = fr_absoluta_engsoftob.div(fr_absoluta_engsoftob.sum(1), axis=0) * 100
pc_engsoftob
```

**R:A porcentagem de Engenheiros de Software com Obesidade é de 25%**

8. Quem tem maior pressão sanguínea média, homens ou mulheres?
(Considere a média)

```python
#separando valores de pressão sistólica e diastólica pelo delimitador
pa = df['Pressão sanguínea'].str.split("/")
pas = pa.str.get(0)
pad = pa.str.get(1)

#novas colunas
df['Pressão sistólica'] = pas
df['Pressão diastólica'] = pad

#convertendo para float
df['Pressão sistólica'] = df['Pressão sistólica'].astype(float)
df['Pressão diastólica'] = df['Pressão diastólica'].astype(float)

#filtrando dataframe para mulheres
media_pa = df.groupby('Gênero')[['Pressão sistólica','Pressão diastólica']].mean().round()
media_pa
```

**R: As mulheres possuem uma pressão sanguínea média maior (aproximadamente 130/86) do que os homens (aproximadamente 127/83).**

10. Pessoas com frequências cardíacas acima de 70 dão mais passos que
pessoas com frequência cardíaca menor ou igual a 70? (Use a média)

```python
#filtrando o dataframe para pessoas com frequencia maior que 70
freq_maior70 = df[df['Frequência cardíaca']>70]

#realizando média
cross_freq_passos_maior70 = freq_maior70.groupby('Frequência cardíaca')['Passos diários'].sum().mean().round()
cross_freq_passos_maior70
```

```python
#filtrando o dataframe para pessoas com frequencia menor ou igual a 70
freq_igualmenor70 = df[df['Frequência cardíaca']<=70]

#realizando média
cross_freq_passos_menor = freq_igualmenor70.groupby('Frequência cardíaca')['Passos diários'].sum().mean().round()
cross_freq_passos_menor
```

**R: Pessoas com frequência cardíaca menor ou igual a 70 dão, em média, mais passos (336320) do que pessoas com frequência cardíaca acima de 70 (61493 passos).**
