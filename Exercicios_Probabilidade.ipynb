- qual a probabilidade de selecionar um numero par?

```python
# gerando randomicamente os 1000 numero naturais inteiros
import random

numeros_lista = [random.randint(0, 999999) for x in range(0, 1000)]

print(numeros_lista)

print(len(numeros_lista))
```

```python
# 1 - probabilidade de selecionar um numero par

# 1.1 obter a quantidade de pares

pares_lista = [x for x in numeros_lista if x % 2 == 0]
print(len(pares_lista))

prob_par = len(pares_lista) / len(numeros_lista)

print(f"A probabilidade de selecionar um número par é: {prob_par}")
```

```python
# 2 - probabilidade de selecionar um numero menor que 2500

# 2.1 filter os numero que sejam menores que 2500

menores_2500_lista = [x for x in numeros_lista if x < 2500]

print(len(menores_2500_lista))

prob_menor = len(menores_2500_lista) / len(numeros_lista)

print(f"A probabilidade de selecionar um número menor que 50 é: {prob_menor}")
```

```python
# 3 - probabilidade de selecionar um numero que nao seja primo

# 3.1 obter a quantidade de numeros primos na amostra

from sympy import *

primo_lista = [x for x in numeros_lista if isprime(x)]

print(len(primo_lista))

prob_primo = len(primo_lista) / len(numeros_lista)

prob_nao_primo = 1 - prob_primo

print(f"A probabilidade de selecionar um número nao primo é: {prob_nao_primo}")
```

```python
# 4 - probabilidade de selecionar um numero divisivel por 11, por 3, e por 7

# 4.1 obter a quantidade de numeros divisiveis por 11, por 3, e por 7

divisivel_lista = [x for x in numeros_lista if x % 11 == 0 and x % 3 == 0 and x % 7 == 0]

print(len(divisivel_lista))

print(divisivel_lista)

prob_divisivel = len(divisivel_lista) / len(numeros_lista)

print(f"A probabilidade de selecionar um número divisivel por 11, por 3, e por 7 é: {prob_divisivel}")
```

```python
# 5. probabilidade de ao selecionar 100 numeros de dentro da nossa amostra de forma aleatoria, 5 deles serem primos

# usar a binimial para esse caso

from scipy.stats import binom

n = 100  # numero de tentativas
k = 5   # numero de sucessos
p = prob_primo  # probabilidade do sucesso

# Calculate the binomial probability
binomial_prob = binom.pmf(k, n, p)

print(f"A probabilidade de ao selecionar 100 numeros de dentro da nossa amostra de forma aleatoria, 5 deles serem primos é: {binomial_prob}")
```
